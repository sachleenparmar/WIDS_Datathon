```r
library(tidyverse)
```

```r
sales = read_csv("sales.csv")
head(sales)
items = read_csv("items.csv")
head(items)
```

```r
items_mutate <- items %>%
        mutate(profit = price - production_cost)
head(items_mutate)


items_summary <- items %>%
        group_by(item_name) %>%
        summarize(mean_price = mean(price),
               mean_prod_cost = mean(production_cost),
                 mean_profit = mean_price - mean_prod_cost)
items_summary
```

```r
profit <- items_summary$mean_price - items_summary$prod_cost
profit
```
